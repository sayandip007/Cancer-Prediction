```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report,
    roc_curve, precision_recall_curve, average_precision_score
)
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import t
```

```python
import pandas as pd
file_path = '/content/drive/My Drive/LUSCexpfile.csv'
df = pd.read_csv(file_path, sep=';')
df.head()
```

```python
df = df.T
df.head()
```

```python
df.columns = df.iloc[0]
df = df[1:]
df = df.rename(columns = {np.nan:'Class'})
df.columns.name = None
df.head()
```

```python
print("Data size: ")
print(f'Rows: {df.shape[0]}')
print(f'Columns: {df.shape[1]}')
```

```python
df.isnull().sum().any()
```

```python
# Features and Labels
X = df.drop(columns=['Class']).astype(float)
y = df['Class'].map({'normal': 0, 'tumor': 1})  # Encode labels
```

```python
# Welch’s t-test for gene filtering

# Split X into tumor and normal samples
X_tumor = X[y == 1]
X_normal = X[y == 0]

# 1) Compute means & variances
mean_tumor = X_tumor.mean(axis=0)
mean_normal = X_normal.mean(axis=0)
var_tumor = X_tumor.var(axis=0, ddof=1)
var_normal = X_normal.var(axis=0, ddof=1)
n_tumor = len(X_tumor)
n_normal = len(X_normal)

# 2) Welch’s t-statistic
denom = np.sqrt(var_tumor/n_tumor + var_normal/n_normal)
t_stats = (mean_tumor - mean_normal) / denom
t_stats = np.where(np.isfinite(t_stats), t_stats, np.nan)  # handle NaN/inf

# 3) Welch–Satterthwaite degrees of freedom
num = (var_tumor/n_tumor + var_normal/n_normal)**2
den = ((var_tumor/n_tumor)**2)/(n_tumor - 1) + ((var_normal/n_normal)**2)/(n_normal - 1)
df_welch = num / den
df_welch = np.where(np.isfinite(df_welch), df_welch, np.nan)

# 4) Two-sided p-values
from scipy.stats import t
p_values = 2 * t.sf(np.abs(t_stats), df_welch)

# 5) Build results DataFrame
pval_df = pd.DataFrame({
    "Gene": X.columns,
    "pval": p_values,
    "mean_expr": X.mean(axis=0).values
})

# 6) Filter significant genes
significant_genes = pval_df.loc[
    (pval_df["pval"] < 0.05) & (pval_df["mean_expr"] > 0.5),
    "Gene"
]

# 7) Debug info
print("Number of genes before filtering:", len(pval_df))
print("Number of significant genes:", len(significant_genes))
print("Top significant genes:\n", significant_genes.head(10))
```

```python
# Filter dataset to significant genes only
X_filtered = X[significant_genes]

# Split into train/test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_filtered, y, test_size=0.35, stratify=y, random_state=42
)
```

```python
# Scale feature
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Ensemble Learning (SVM + RF)
svm_clf = SVC(
    kernel='rbf',
    probability=True,
    C=10,
    gamma='scale',
    random_state=42
)

rf_clf = RandomForestClassifier(
    n_estimators=200,       # number of trees
    max_depth=None,         # let trees expand fully
    random_state=42,
    n_jobs=-1,
    class_weight='balanced' # handle imbalance if any
)

# Soft voting allows probability averaging
ensemble_clf = VotingClassifier(
    estimators=[('svm', svm_clf), ('rf', rf_clf)],
    voting='soft'
)

ensemble_clf.fit(X_train_scaled, y_train)
y_pred = ensemble_clf.predict(X_test_scaled)
y_prob = ensemble_clf.predict_proba(X_test_scaled)[:, 1]
```

```python
# Evaluation Metrics
print("✅ Ensemble Model Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_prob))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Normal", "Tumor"],
            yticklabels=["Normal", "Tumor"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (Ensemble)")
plt.show()
```

```python
# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, y_prob):.2f}")
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve (Ensemble)")
plt.legend()
plt.show()
```

```python
# Classification Report
print(classification_report(y_test, y_pred, target_names=["Normal", "Tumor"]))
```

```python
# Precision-Recall Curve
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
ap = average_precision_score(y_test, y_prob)
plt.plot(recall, precision, label=f"AP = {ap:.2f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (Ensemble)")
plt.legend()
plt.show()
```

```python
# Learning Curve (Bias vs Variance Check)
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(
    ensemble_clf, X, y, cv=5, scoring='accuracy', n_jobs=-1
)

train_mean = train_scores.mean(axis=1)
test_mean = test_scores.mean(axis=1)

plt.plot(train_sizes, train_mean, label="Training score")
plt.plot(train_sizes, test_mean, label="Validation score")
plt.xlabel("Training Set Size")
plt.ylabel("Accuracy")
plt.title("Learning Curve")

plt.legend()
plt.show()
```
