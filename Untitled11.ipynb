```python
# ============================================================
# REFINED PROJECT: RANDOM FOREST (RF) PRIORITIZATION OF CANCEROUS BIOMARKERS
# (Final Code with Gene Name Extraction)
# ============================================================

import pandas as pd
import numpy as np
from statsmodels.stats.weightstats import ztest
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
import warnings, re
from collections import defaultdict

warnings.filterwarnings("ignore")

# --- DATA LOADING (Corrected Filtering) ---
file_path = '/content/drive/My Drive/gene_expression_data_for_Lung.xlsx'
df_raw = pd.read_excel(file_path)
df = df_raw.iloc[2:].reset_index(drop=True)

cols_to_drop = [c for c in ['Target', 'Unnamed: 1'] if c in df.columns]
df.drop(columns=cols_to_drop, inplace=True)
if 'Unnamed: 0' in df.columns:
    df.rename(columns={'Unnamed: 0': 'GENE'}, inplace=True)

# --- CRITICAL FIX: Ensure only valid gene rows remain ---
df.dropna(subset=['GENE'], inplace=True)
df.set_index('GENE', inplace=True)

normal_columns = [col for col in df.columns if col.startswith("LN")]
cancer_columns = [col for col in df.columns if (col.startswith("AD") or (col.startswith("L") and not col.startswith("LN")))]

df = df.apply(pd.to_numeric, errors='coerce').fillna(0)

# =======================================
# STEP 1: CALCULATE DIFFERENTIAL EXPRESSION METRICS
# =======================================
print("Calculating Fold Change and P-values...")
deg_metrics = []
for index, row in df.iterrows():
    cancer_vals = row[cancer_columns].astype(float)
    normal_vals = row[normal_columns].astype(float)
    mean_cancer = cancer_vals.mean()
    mean_normal = normal_vals.mean()
    FC = (mean_cancer + 1) / (mean_normal + 1)
    log2_FC = np.log2(FC)
    try:
        z_score, p_val = ztest(cancer_vals, normal_vals, value=0)
    except:
        p_val = 1.0
    deg_metrics.append({'GENE': index, 'log2_FC': log2_FC, 'p_value': p_val})

deg_df = pd.DataFrame(deg_metrics).set_index('GENE')
df = df.join(deg_df)
df["is_DE"] = (df["p_value"] < 0.05).astype(int)

# =======================================
# STEP 2: DEFINE TARGET VARIABLE (Y) FROM REFERENCE LIST
# =======================================
alias_file = '/content/drive/My Drive/lung_cancer_genes1.txt'
with open(alias_file, 'r', encoding='utf-8') as f: raw_text = f.read()
raw_text = raw_text.replace("–", "-").replace("—", "-")
lines = [line.strip() for line in raw_text.split("\n") if line.strip()]
gene_map = defaultdict(list)
for line in lines:
    if ":" in line or "\t" in line:
        parts = re.split(r'[:\t]', line, maxsplit=1)
        main = parts[0].strip().upper()
        aliases = [a.strip().upper() for a in re.split(r'[,;/ ]+', parts[1]) if a.strip()]
        gene_map[main].extend(aliases)
alias_to_main = {alias: main for main, aliases in gene_map.items() for alias in aliases}
known_genes = set(gene_map.keys())
df.index = df.index.astype(str)
def map_to_main(g):
    g = str(g).upper()
    if g in gene_map: return g
    if g in alias_to_main: return alias_to_main[g]
    return g
df["is_known_cancer_gene"] = [1 if map_to_main(g) in known_genes else 0 for g in df.index]
y_new = df["is_known_cancer_gene"]


# =======================================
# STEP 3: RANDOM FOREST CLASSIFIER
# =======================================
X_new = df[cancer_columns + normal_columns]
y_new = df["is_known_cancer_gene"]

print(f"Class Distribution (Known Genes): {y_new.sum()} known out of {len(y_new)} total.")

X_train, X_test, y_train, y_test = train_test_split(
    X_new, y_new, test_size=0.3, random_state=42, stratify=y_new
)

# Using optimized parameters from previous successful runs
rf_model = RandomForestClassifier(n_estimators=300,
                                  random_state=42,
                                  class_weight='balanced',
                                  max_depth=None,
                                  min_samples_leaf=5,
                                  min_samples_split=5)

rf_model.fit(X_train, y_train)

# -------- FEATURE IMPORTANCE PLOT --------
print("\nAnalyzing Feature (Sample) Importance...")
importances = rf_model.feature_importances_
feature_names = X_new.columns
forest_importances = pd.Series(importances, index=feature_names).sort_values(ascending=False)

plt.figure(figsize=(14, 6))
forest_importances.head(20).plot.bar(color='#1e8546')
plt.title("Top 20 Most Important Sample Features for Gene Classification", fontsize=14, fontweight='bold')
plt.ylabel("Importance (Gini Index Decrease)")
plt.xlabel("Sample ID (Feature)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

y_pred = rf_model.predict(X_test)

print("\nModel Training Completed Successfully with Random Forest! ")
print("\nTEST PERFORMANCE (Predicting Known Genes):")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred, zero_division=0):.4f}")
print(f"Recall: {recall_score(y_test, y_pred, zero_division=0):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred, zero_division=0):.4f}")

# =======================================
# STEP 4: IDENTIFY CANDIDATE CANCEROUS GENES
# =======================================
df["predicted_cancer_gene"] = rf_model.predict(X_new)

candidate_genes_df = df[df["predicted_cancer_gene"] == 1]
candidate_genes_df.to_csv("candidate_cancer_genes_RF.csv", index=True)

print("\nCandidate Cancer Genes Identified and Saved to 'candidate_cancer_genes_RF.csv'")
print(f"Total Candidates Predicted: {len(candidate_genes_df)}")

# ============================================================
# VALIDATION: Assessing Biological Overlap (with Gene Name Extraction)
# ============================================================

predicted_genes_canon = set(map_to_main(g) for g in candidate_genes_df.index)
known_genes_canon = known_genes
all_genes_canon = set(map_to_main(g) for g in df.index.astype(str))

# Calculate Overlap Metrics
TP_genes = predicted_genes_canon.intersection(known_genes_canon)
FP_genes = predicted_genes_canon.difference(known_genes_canon)
FN_genes = known_genes_canon.difference(predicted_genes_canon)
TN_genes = all_genes_canon.difference(predicted_genes_canon.union(known_genes_canon))

TP, FP, FN, TN = len(TP_genes), len(FP_genes), len(FN_genes), len(TN_genes)
precision = (TP / (TP + FP) * 100) if TP + FP else 0
recall = (TP / (TP + FN) * 100) if TP + FN else 0
f1 = (2 * precision * recall / (precision + recall)) if (precision + recall) else 0

print("\nFINAL GENE VALIDATION AGAINST REFERENCE LIST")
print("---------------------------------------------")
print(f"True Positives (TP): {TP} (Predicted and Known Cancer Genes)")
print(f"False Positives (FP): {FP} (Predicted Cancer, but NOT Known)")
print(f"False Negatives (FN): {FN} (Known Cancer, but NOT Predicted)")
print(f"True Negatives (TN): {TN} (Neither Predicted nor Known Cancer)")
print("\nValidation Metrics:")
print(f"Precision (How accurate are my predictions?): {round(precision, 2)}%")
print(f"Recall (How many known genes did I find?): {round(recall, 2)}%")
print(f"F1 Score: {round(f1, 2)}")

# ---  NEW: GENE NAME EXTRACTION  ---
print("\n HIGH-CONFIDENCE PREDICTED CANCER GENES ")
print("--------------------------------------------------")

# 1. True Positive Gene Names (Confirmed Biomarkers)
print(f"True Positive Genes (Confirmed Biomarkers - {TP} genes):")
if TP > 0:
    print(f"[{TP} genes]: " + ", ".join(sorted(list(TP_genes))))
else:
    print("No True Positives identified in the predicted set.")

# 2. False Positive Gene Names (Novel Candidates)
print(f"\n False Positive Genes (NOVEL CANDIDATES - {FP} genes):")
if FP > 0:
    print(f"[{FP} genes]: " + ", ".join(sorted(list(FP_genes))))
else:
    print("No False Positives identified. The model achieved near-perfect purity.")

# Visualization (Confusion Matrix)
conf_matrix = np.array([[TP, FP], [FN, TN]])

plt.figure(figsize=(8, 6))
colors = ["#073b20","#1e8546","#77DD77","#64E986","#a9ebbd"]
cmap = LinearSegmentedColormap.from_list("CustomCmap", colors)
sns.heatmap(conf_matrix, annot=True, fmt=".0f", cmap=cmap,
            xticklabels=["Predicted Cancer", "Predicted Non-Cancer"],
            yticklabels=["Actual Cancer", "Actual Non-Cancer"],
            cbar_kws={'label': 'Gene Count'})
plt.title("Gene-Level Confusion Matrix (Random Forest Validation)", fontsize=14, fontweight='bold')
plt.xlabel("Prediction Outcome")
plt.ylabel("Actual Status (Reference List)")
plt.tight_layout()
plt.show()

```
