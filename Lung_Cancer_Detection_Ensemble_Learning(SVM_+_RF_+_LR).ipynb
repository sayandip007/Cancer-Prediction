```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report,
    roc_curve, precision_recall_curve, average_precision_score
)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
file_path = '/content/drive/My Drive/LUSCexpfile.csv'
df = pd.read_csv(file_path, sep=';')
df.head()
```

```python
df = df.T
df.head()
```

```python
df.columns = df.iloc[0]
df = df[1:]
df = df.rename(columns = {np.nan:'Class'})
df.columns.name = None
df.head()
```

```python
print("Data size: ")
print(f'Rows: {df.shape[0]}')
print(f'Columns: {df.shape[1]}')
```

```python
df.isnull().sum().any()
```

```python
# Features and Labels
X = df.drop(columns=['Class']).astype(float)
y = df['Class'].map({'normal': 0, 'tumor': 1})  # Encode labels
```

```python
# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.35, stratify=y, random_state=42
)
```

```python
# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Ensemble Learning
log_reg = LogisticRegression(
    penalty='l1',
    C=1.0,
    solver='liblinear',
    max_iter=1000,
    random_state=42
)

svm_clf = SVC(
    kernel='rbf',
    probability=True,
    C=10,
    gamma='scale',
    random_state=42
)

rf_clf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42,
    n_jobs=-1,
    class_weight='balanced'
)

# Stacking Classifier
stack_clf = StackingClassifier(
    estimators=[('lr', log_reg), ('svm', svm_clf), ('rf', rf_clf)],
    final_estimator=LogisticRegression(max_iter=1000, random_state=42),
    passthrough=True,  # pass original features + base model outputs
    cv=5,              # cross-validation for meta-learner
    n_jobs=-1
)

# Train & Predict
stack_clf.fit(X_train_scaled, y_train)
y_pred = stack_clf.predict(X_test_scaled)
y_prob = stack_clf.predict_proba(X_test_scaled)[:, 1]
```

```python
# Evaluation Metrics
print("✅ Stacking Ensemble Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_prob))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Normal", "Tumor"],
            yticklabels=["Normal", "Tumor"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (Stacking Ensemble)")
plt.show()
```

```python
# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, y_prob):.2f}")
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve (Stacking Ensemble)")
plt.legend()
plt.show()
```

```python
# Classification Report
print(classification_report(y_test, y_pred, target_names=["Normal", "Tumor"]))
```

```python
# Precision-Recall Curve
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
ap = average_precision_score(y_test, y_prob)
plt.plot(recall, precision, label=f"AP = {ap:.2f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (Stacking Ensemble)")
plt.legend()
plt.show()
```

```python
# Learning Curve (Bias vs Variance Check)
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(
    stack_clf, X, y, cv=5, scoring='accuracy', n_jobs=-1
)

train_mean = train_scores.mean(axis=1)
test_mean = test_scores.mean(axis=1)

plt.plot(train_sizes, train_mean, label="Training score")
plt.plot(train_sizes, test_mean, label="Validation score")
plt.xlabel("Training Set Size")
plt.ylabel("Accuracy")
plt.title("Learning Curve")

plt.legend()
plt.show()
```
